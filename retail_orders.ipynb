## Que: `find top 5 highest revenue generating products`

```python
pd.read_sql_query('''
                select product_id, sum(sale_price) as sales 
                from retail_orders
                group by product_id
                order by sales desc
                limit 5
                ''',conn)
```

## Que: `find top 5 highest selling products in each region`
-----------

```python
pd.read_sql_query(''' select region, product_id, sum(sale_price) as sales
                from retail_orders
                group by region, product_id
                order by region, sales desc
                limit 5
                 ''',conn)
```

## Que: `Find month over month growth comparison for 2022 and 2023 sales eg: jan 2022 vs jan 2023`
--------------

```python
pd.read_sql_query(''' with cte as (
                select year(order_date) as order_year,
                month(order_date) as order_month,
                sum(sale_price) as sales
                from retail_orders
                group by year(order_date), month(order_date)
                )
                  
                select order_month,
                sum(case when order_year=2022 then sales else 0 end) as sales_2022,
                sum(case when order_year=2023 then sales else 0 end) as sales_2023
                from cte
                group by order_month
                order by order_month
                ''',conn)
```

--------------------------
## Que: `For each category which month had highest sales`
--------------------------

```python
pd.read_sql_query(''' with cte as (
                select category, format(order_date,'yyyymm') as order_year_month,
                sum(sale_price) as sales
                from retail_orders
                group by category, format(order_date,'yyyymm')
                ) 
                
                select *from (
                select *,
                row_number() over(partition by category order by sales desc) as rn
                from cte
                ) a where rn = 1
                ''',conn)
```

------------------
## Que: `Which sub category had highest growth by profit(in percentage%) in 2023 compare to 2022`
------------------

```python
pd.read_sql_query(
    '''
    with cte as (
    select sub_category, year(order_date) as order_year,
    sum(sale_price) as sales
    from retail_orders
    group by sub_category, year(order_date)
    ),

    cte2 as (
    select sub_category,
    sum(case when order_year=2022 then sales else 0 end) as sales_2022,
    sum(case when order_year=2023 then sales else 0 end) as sales_2023
    from cte
    group by sub_category
    )

    select *,
    (sales_2023 - sales_2022) * 100 / sales_2022
    from cte2
    order by (sales_2023 - sales_2022) * 100 / sales_2022 desc
''',conn
)
```
